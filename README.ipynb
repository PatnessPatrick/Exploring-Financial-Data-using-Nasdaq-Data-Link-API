# Exploring Financial Data using Nasdaq Data Link API

## Project Overview

This project explores the financial data provided by the Nasdaq Data Link API, focusing on the MER/F1 datatable. The primary aims were to analyze the financial trends of 'accrued expenses turnover' over time for various companies and to examine the contributions of different geographical regions to these financial metrics.

## Objectives

1. **Analyze Financial Trends Over Time**:
   - Investigate the 'accrued expenses turnover' for various companies from 2010 to 2015.
   
2. **Geographical Regional Analysis**:
   - Assess the performance of different regions in terms of accrued expenses turnover.

## Methodology

### Data Collection

Utilized the Nasdaq Data Link API to extract detailed financial data, including balance sheets, cash flow statements, and income statements for various companies.

### Data Processing

1. **Read and Clean Data**:
   - Loaded the data into a pandas DataFrame.
   - Renamed columns to follow snake_case conventions.
   - Mapped country abbreviations to full country names.
   - Converted the `report_date` column to datetime format.
   
2. **Financial Trend Analysis**:
   - Filtered data for the period 2010-2015.
   - Visualized time series trends for each companyâ€™s accrued expenses turnover.

3. **Geographical Regional Analysis**:
   - Grouped data by country to calculate average accrued expenses turnover.
   - Visualized the average accrued expenses turnover for each country.

## Key Findings

### Financial Trends Over Time

The analysis of financial trends for various companies from 2010 to 2015 revealed:

- **Company-Specific Trends**: Companies like Apple Inc and Accenture plc exhibited stable trends, indicating consistent financial management practices. UltraPetrol (Bahamas) showed significant variations, suggesting operational changes or external factors.
- **Overall Trends**: Most companies showed an upward trend in accrued expenses turnover, reflecting improved efficiency in managing short-term liabilities.

### Geographical Regional Analysis

The geographical analysis highlighted significant differences in financial practices and efficiency across regions:

- **Top Performing Regions**:
  - **Bahamas**: Highest average accrued expenses turnover (66.52), indicating strong cash flow management.
  - **United States**: Second highest (44.38), reflecting robust financial health.
  - **United Kingdom**: Similar performance to the United States (44.17).
  
- **Moderate to Lower Performing Regions**:
  - **Cayman Islands**: Moderate efficiency (15.06).
  - **Japan, Germany, and Ireland**: Lower average turnover rates (10.38, 9.29, and 7.26, respectively), suggesting more conservative financial management practices.
